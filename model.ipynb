```python
import plotly.graph_objects as go
#from utils_plot import *
from kaleido import *
import matplotlib.pyplot as plt
import numpy as np

```

```python
def add_optimum(ax, point, text):

        ax_point = ax.axvline(point, 0,1.5, color='k', lw = 0.5)
        trans = ax.transData.transform(ax_point.get_xydata())
        trans = ax.transAxes.inverted().transform(trans)
        #ax.text(trans[1][0]+0.01,0.6,
        #        s=text,fontsize = 'xx-small',
        #        rotation=90, transform=ax.transAxes)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1, figsize=(7.50, 4.50),dpi =300)

delta = 0.1
xrange = np.arange(0.1, 200.0, delta)

x = 2500
equation = xrange * (16 * (1 - xrange/x ) - 15)
optimum = x * (16-15) /  ( 2 * 16)
add_optimum(ax, optimum, 'opt')
ax.plot(xrange, equation)

x = 2000
equation2 = xrange * (16 * (1 - xrange/x ) - 15)
optimum = x * (16-15) /  ( 2 * 16)
add_optimum(ax, optimum, 'opt')
ax.plot(xrange, equation2)

x = 3000
equation2 = xrange * (16 * (1 - xrange/x ) - 15)
optimum = x * (16-15) /  ( 2 * 16)
add_optimum(ax, optimum, 'opt')
ax.plot(xrange, equation2)

#ax.show()
```
